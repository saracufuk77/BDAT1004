```python
#Assignment 5 Part 3

import sqlite3

conn = sqlite3.connect('group7.sqlite')
cur = conn.cursor()

cur.execute('DROP TABLE IF EXISTS Students')
cur.execute('CREATE TABLE Students (stu_name TEXT, stu_id INTEGER)')
cur.execute('DROP TABLE IF EXISTS Courses')
cur.execute('CREATE TABLE Courses (crs_name TEXT, crs_id INTEGER)')
cur.execute('DROP TABLE IF EXISTS Relation')
cur.execute('CREATE TABLE Relation (stu_id INTEGER, crs_id INTEGER)')
conn.commit()

cur.executemany("INSERT INTO Students (stu_name, stu_id) VALUES (?,?)",[("Ufuk", 1),("Cihat",2),("Abdulkadir",3),("Mehmet Ali",4)])
cur.executemany("INSERT INTO Courses (crs_name,crs_id) VALUES (?,?)",[("Data Manipulation",1000),("Information Encoding",1001),("System Architecture",1002),("Business Process",1003),("Programing", 1004),("Mathematics",1005)])
cur.executemany("INSERT INTO Relation (stu_id, crs_id) VALUES (?,?)",[(1,1004),(1,1005),(1,1000),(1,1002),(2,1004),(2,1003),(2,1001),(3,1003),(3,1000),(3,1002),(4,1000),(4,1004)])

conn.commit()

print("Students table:")            
cur.execute('SELECT stu_name, stu_id FROM Students')
for row in cur:
    print(row)

print("\nCourses table:")
cur.execute('SELECT crs_name, crs_id FROM Courses')
for row in cur:
    print(row)

print("\nRelations:")
cur.execute('SELECT Students.stu_id, Students.stu_name, Relation.crs_id, Courses.crs_name From Students JOIN Relation ON Students.stu_id=Relation.stu_id JOIN Courses ON Relation.crs_id=Courses.crs_id')
for row in cur:
    print(row)
    
conn.close()
```
